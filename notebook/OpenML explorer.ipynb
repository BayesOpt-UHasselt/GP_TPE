```python
from sklearn.datasets import fetch_openml
from openml import datasets, tasks, runs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataset = fetch_openml(name='banana')

X = np.asarray(dataset.data)
Y = np.asarray(dataset.target)
```

```python
df = pd.DataFrame(dict(y = Y))

fig, ax = plt.subplots(figsize=(10,10))

colors = {'1':'red', '2':'blue'}


ax.scatter(X[:,0], X[:,1], c=df["y"].map(colors))
```
