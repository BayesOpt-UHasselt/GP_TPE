```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
matrix =  np.random.rand(30, 1)
a = np.zeros(10)
a = np.append(a, np.ones(10))
a = np.append(a, 1+np.ones(10))

b = np.zeros(5)
for j in range(1, 6):
    if j%2 == 0:
        b = np.append(b, np.zeros(5))
    else:
        b = np.append(b, np.ones(5))
        
df = pd.DataFrame(matrix)
df["3"] = a
df["4"] = b
```

```python
df
```

```python
medians = np.zeros((3, 2))
matrix = df.values
for i in range(medians.shape[0]):
    factor1 = matrix[matrix[:, 1]==i,:]
    for j in range(medians.shape[1]):
        factor2 = factor1[factor1[:, 2]==j,:]
        m = np.median(factor2, axis=0)
        medians[i, j] = m[0]
        print(factor2)
        print(m)
        print()

fig, ax = plt.subplots(figsize=(5,10))
colors = {0.0:'red', 1:'blue'}
for i in range(3):
    ax.scatter(a[a==i], matrix[a==i,0], c=df["4"][a==i].map(colors))

for m in range(medians.shape[1]):
    plt.plot(medians[:, m], color=colors[m])
```
